```python
import os
import pandas as pd
from tqdm import tqdm
import spacy

def construct_csv(fname):
    example = list()
    with open(fname, 'r') as f:
        lines = f.readlines()
        topic = list()
        old = list()
        new = list()
        ctx = list()
        for line in tqdm(lines):
            if line[0] == '=':
                topic = line.strip('=').strip('=\n').strip()
            elif line[:3] ==  'OLD':
                old = line.strip('OLD: ').strip()
            elif line[:3] ==  'NEW':
                new = line.strip('NEW: ').strip()
            elif '----' in line:
                example.append([topic, old, new, ctx])
                topic = list()
                old = list()
                new = list()
                ctx = list()
            elif line != '':
                ctx = line.strip('\t').strip()
            else:
                print(line)
    return example
```

```python
fname = './wikiHow_revisions_filtered_dev1perc_wContext.txt'
data = construct_csv(fname)
data[1]
```

```python
df = pd.DataFrame(data, columns = ['Topic', 'Old', 'New', 'Context'])
df
```

```python
for i in range(len(df['Old'])):
    print(str(df['Old'][i]) + '\n' + str(df['New'][i]))
    print('\n')
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")

t0 = nlp(df['Old'][1])
t1 = nlp(df['New'][1])

print([t for t in t0])
print([t.tag_ for t in t0])

print([t for t in t1])
print([t.tag_ for t in t1])
```

```python
old_pos = list()
new_pos = list()
old_dep = list()
new_dep = list()
for ix in tqdm(range(len(df['Old']))):
    try:
        t0 = nlp(df['Old'][ix])
        t1 = nlp(df['New'][ix])
    except Exception as e:
        t0 = nlp('')
        t1 = nlp('')
        print(df['Old'][ix], df['New'][ix])
        print(e)
    old_pos.append(' '.join([t.tag_ for t in t0]))
    new_pos.append(' '.join([t.tag_ for t in t1]))
    old_dep.append(' '.join([t.dep_ for t in t0]))
    new_dep.append(' '.join([t.dep_ for t in t1]))

df['Old POS'] = old_pos
df['Old Dep'] = old_dep
df['New POS'] = new_pos
df['New Dep'] = new_dep
```

```python
flag = 0
for ix in range(len(df['Topic'])):
    if type(df['Old'][ix]) != str:
        print(df['Old'][ix], df['New'][ix])
    elif len(df['Old'][ix].split()) == len(df['New'][ix].split()):
        t0 = nlp(df['Old'][ix])
        t1 = nlp(df['New'][ix])
        t0_tags = [t.tag_ for t in t0]
        t1_tags = [t.tag_ for t in t1]
        t0_dep = [t.dep_ for t in t0]
        t1_dep = [t.dep_ for t in t1]
        for t in range(len(t0_tags)):
            try:
                if str(t0[t]) != str(t1[t]):
                    print('Word', end='\t')
                    print(str(t0).split()[t], t0_tags[t], t0_dep[t], end='\t\t')
                    print(str(t1).split()[t], t1_tags[t], t1_dep[t])
                if t0_tags[t] != t1_tags[t]:
                    print('POS Tags', end='\t')
                    print(str(t0).split()[t], t0_tags[t], t0_dep[t], end='\t\t')
                    print(str(t1).split()[t], t1_tags[t], t1_dep[t])
                    flag = 1
                if t0_dep[t] != t1_dep[t]:
                    print('Dependency', end='\t')
                    print(str(t0).split()[t], t0_tags[t], t0_dep[t], end='\t\t')
                    print(str(t1).split()[t], t1_tags[t], t1_dep[t])
                    flag = 1
            except:
                print('Exception:')
                print(t0)
                print(' '.join(t0_tags))
                print(t1)
                print(' '.join(t1_tags))
                print('\n')
        if flag == 1:
            print(t0)
            print(' '.join(t0_tags))
            print(' '.join(t0_dep))
            print(t1)
            print(' '.join(t1_tags))
            print(' '.join(t1_dep))
            print('\n')
            flag = 0
```
